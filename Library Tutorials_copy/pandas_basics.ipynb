```python
newdf3=pd.DataFrame({"name" : ['Hrithik','Ragini','Abhishek','Radha','xyz'],
                     "DOBmonth" : ['Jan','Sept','Nov','Jul', np.nan],
                     "DOByear" : [1998,1997,2003,1971,pd.NaT]
                    })
newdf3
```

```python
newdf3.dropna()    #drops the column where even a single None is present
```

```python
newdf3.dropna(how='all')    #it will remove the row only if all the values are None in that row
```

```python
newdf3=pd.DataFrame({"name" : ['Hrithik','Ragini','Abhishek','Radha','xyz'],
                     "DOBmonth" : ['Jan','Sept','Nov','Jul', np.nan],
                     "DOByear" : [1998,1997,2003,1971,pd.NaT],
                     "DOBdate" : [pd.NaT,pd.NaT,pd.NaT,pd.NaT,pd.NaT]
                    })
newdf3
```

```python
newdf3.dropna(how='all',axis=1)    #to remove a whole None column
```

```python
newdf3=pd.DataFrame({"name" : ['Hrithik','Ragini','Abhishek','xyz','xyz'],
                     "DOBmonth" : ['Jan','Sept','Nov','Jul', np.nan],
                     "DOByear" : [1998,1997,2003,1971,pd.NaT],
                     "DOBdate" : [pd.NaT,pd.NaT,pd.NaT,pd.NaT,pd.NaT]
                    })
newdf3
```

```python
newdf3.drop_duplicates(subset=['name'])   #it doesn't take axis as input hence the subset, by default it keeps the 1st duplicate
```

```python
newdf3.drop_duplicates(subset=['name'],keep='last')    #if we want to keep only the last duplicate
```

```python
newdf3.drop_duplicates(subset=['name'],keep=False)   #if we don't want to keep any of the duplicates
```
