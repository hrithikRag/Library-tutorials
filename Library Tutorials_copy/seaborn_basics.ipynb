# SEABORN TUTORIAL

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

**LINE PLOT :**

```python
fmri=sns.load_dataset("fmri")
fmri
```

```python
sns.lineplot(x="timepoint", y="signal", data=fmri)
plt.show()
```

```python
sns.lineplot(x="timepoint", y="signal", hue="event" , data=fmri)
plt.show()    #if we want line plot of different attributes that are present in dataset
```

```python
sns.lineplot(x="timepoint", y="signal", hue="event", style="event", data=fmri, markers=True)
plt.show()   #style for different type of lines(solid,dashed), markers for keeping points on the line plot
```

**BAR PLOT :**

```python
pokemon=pd.read_csv('pokemon.csv')    #downloaded it from kaggle in csv format
pokemon.head()
```

```python
sns.set(style='whitegrid')
sns.barplot(x='is_legendary',y='speed',data=pokemon)
plt.title('speed vs status')
plt.show()
```

```python
sns.set(style='whitegrid')
sns.barplot(x='is_legendary',y='weight_kg',data=pokemon)
plt.title('weight vs status')
plt.show()
```

```python
sns.set(style='whitegrid')
sns.barplot(x='is_legendary',y='speed',hue='generation', data=pokemon)
plt.title('speed vs status')
plt.show()
```

```python
sns.set(style='whitegrid')
sns.barplot(x='is_legendary',y='weight_kg',hue='generation', data=pokemon)
plt.title('weight vs status')
plt.show()
```

```python
sns.set(style='whitegrid')
sns.barplot(x='is_legendary',y='weight_kg',palette='Blues_d',data=pokemon)
plt.title('weight vs status')
plt.show()    #similarly there are different palettes available for different shades
```

```python
sns.set(style='whitegrid')
sns.barplot(x='is_legendary',y='weight_kg',color='orange',data=pokemon)
plt.title('weight vs status')
plt.show()    #if we want to assign same color to both bar
```

**SCATTER PLOT :**

```python
sns.scatterplot(x='pokedex_number', y='weight_kg',hue='generation',style='is_legendary',data=pokemon)
plt.show()    #hue is for different colour and style is for using different symbol for different attributes
```

```python
sns.scatterplot(x='pokedex_number', y='weight_kg',hue='is_legendary',style='generation',data=pokemon)
plt.show()    #if interchange the two above setting
```

**HISTOGRAM :**

```python
plt.subplot(1,2,1)
sns.histplot(pokemon['is_legendary'],bins=2)    #by default the smooth curve will not be shown

plt.subplot(1,2,2)
sns.histplot(pokemon['generation'], color='red',bins=7, kde=True)    #we have to use kde if we also want smooth curve
plt.show()
```

**JOINT PLOT :**

```python
sns.jointplot(x='pokedex_number', y='weight_kg', color='olive',kind='reg', data=pokemon)  #reg for regression line
```

**BOX PLOT :**

```python
sns.boxplot(x='generation', y='weight_kg', data=pokemon, linewidth=1)   #linewidth is to adjust the width of outlines
plt.show()
```

```python
sns.boxplot(x='generation', y='speed', data=pokemon, palette='Set1', order=[7,6,5,4,3,2,1], hue='is_legendary', linewidth=2)
plt.show()    #order is used to set attributes along x-axis in any random order
```
