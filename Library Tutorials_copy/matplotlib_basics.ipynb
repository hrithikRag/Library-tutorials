# Matplotlib tutorial

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

**LINE PLOT :**

```python
x=np.arange(1,11)
x
```

```python
y=x**2
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(y,x)
plt.show()
```

```python
plt.plot(x,y,color='orange',linestyle=':', linewidth=10)
plt.title("First line plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(x,y)       #if we want to join dots again with a line
plt.show()
```

```python
y1=x**3
y1
```

```python
plt.plot(x,y,color='orange',linestyle=':', linewidth=4)
plt.plot(x,y1,color='r',linestyle=':', linewidth=4)
plt.title("Second line plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

```python
plt.subplot(1,2,1)     #1 row, 2 column, 1st element
plt.plot(x,y,color='orange',linestyle=':', linewidth=4)

plt.subplot(1,2,2)      #1 row, 2 column, 2nd element
plt.plot(x,y1,color='r',linestyle=':', linewidth=4)

plt.show()
```

```python
plt.subplot(2,1,1)      #2 row, 1 column, 1st element
plt.plot(x,y,color='orange',linestyle=':', linewidth=4)

plt.subplot(2,1,2)       #2 row, 1 column, 2nd element
plt.plot(x,y1,color='r',linestyle=':', linewidth=4)

plt.show()
```

**BAR PLOT :**

```python
student={'Peter':87,'Alfred':45,'Bruce':97,'Tony':100,'Steve':95,'Thor':70,'Natasha':90,'Scot':50,'Steven':97}
names=list(student.keys())
values=list(student.values())
plt.bar(names,values)
plt.title("First bar plot")
plt.xlabel("NAME")
plt.ylabel("MARKS")
plt.grid(True)
plt.show()
```

```python
plt.barh(names,values,color='r')      #for horizontal bar plot
plt.title("First bar plot")
plt.xlabel("NAME")
plt.ylabel("MARKS")
plt.grid(True)
plt.show()
```

**SCATTER PLOT :**

```python
ser2=list(pd.Series(np.random.rand(100)))
ser1=list(np.arange(0,100))
plt.scatter(ser1,ser2)
plt.show()
```

```python
plt.scatter(ser1,ser2,marker='*',c='orange',s=50)    #s stands for size
plt.show()
```

```python
ser3=list(pd.Series(np.random.rand(100)))
plt.scatter(ser1,ser2,marker='*',c='orange',s=50)
plt.scatter(ser1,ser3,c='green',s=50)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(ser1,ser2,marker='*',c='orange',s=50)

plt.subplot(1,2,2)
plt.scatter(ser1,ser3,c='green',s=50)

plt.show()
```

```python
plt.subplot(2,1,1)
plt.scatter(ser1,ser2,marker='*',c='orange',s=50)

plt.subplot(2,1,2)
plt.scatter(ser1,ser3,c='green',s=50)

plt.show()
```

**HISTOGRAM :**

```python
data=[0,0,1,1,1,1,2,2,2,3,3,4,4,4,4,4,5,5,6,7,8,8,8,9]
plt.hist(data)
plt.show()
```

**VIOLIN PLOT :**

**PIE CHART :**

```python
fruit=['apple','orange','mango','guava','jamun','leechi']
quantity=[34,54,65,24,88,12]
plt.pie(quantity,labels=fruit)
plt.show()
```

```python
plt.pie(quantity, labels=fruit, autopct="%0.2f%%", colors=['red','orange','yellow','green','purple','pink'])
plt.show()    #autopct to add percentage, 0.1 or 0.2 is no. of decimal places upto which we want the percent 
```

**DOUGH-NUT CHART :**

```python
plt.pie(quantity, labels=fruit, autopct="%0.2f%%", colors=['red','orange','yellow','green','purple','pink'], radius=2)
plt.pie([1],colors=['white'],radius=0.8)   
plt.show()    #we didn't pass label to avoid overlap with percent, whole circle is there because quantity is 1
```
